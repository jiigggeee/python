```python
df_population_raw = pd.read_csv('population_total.csv')
```

```python
df_population_raw
```

```python
df_pivot = df_population_raw.pivot_table(index="year", columns="country", 
                              values="population")
```

```python
df_pivot = df_pivot[['United States', 'India', 'Argentina']]
```

```python
df_pivot
```

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', 
              title='Population (1955-2020)',
             figsize=(8,4))
```

```python
df_pivot
```

```python
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Year', 
                   ylabel='Population',
                  title='Population (2020)')
```
